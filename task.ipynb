```python
import numpy as np
import pandas as pd
```

```python

import warnings
warnings.filterwarnings('ignore')

# Step 1: Load all files
orders = pd.read_csv('List of Orders.csv')
details = pd.read_csv('Order Details.csv')
target = pd.read_csv('Sales target.csv')

print("Shape:", orders.shape, details.shape, target.shape)
```

```python
# Convert Order Date
orders['Order Date'] = pd.to_datetime(orders['Order Date'], dayfirst=True, errors='coerce')

# Check for any parsing issues
print(orders['Order Date'].isnull().sum())
```

```python
# Main merge
df = pd.merge(details, orders, on='Order ID', how='left')

# Optional: Target merge (Category + Month level)
df['Month'] = df['Order Date'].dt.strftime('%b')
df['Year'] = df['Order Date'].dt.year

# Merge target (adjust column names as per your file)
# target = target.rename(columns={'Category': 'Category', 'Target': 'Target Amount'})
```

```python
# Convert numeric columns
numeric_cols = ['Amount', 'Profit', 'Quantity']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Create useful columns
df['Profit Margin %'] = (df['Profit'] / df['Amount'] * 100).round(2)
df['Order Value'] = df['Amount']
```

```python
orders['CustomerName'] = orders['CustomerName'].fillna('Unknown')
orders['State'] = orders['State'].fillna('Unknown')
orders['City'] = orders['City'].fillna('Unknown')
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
# Zone Creation (Very Important for Indian Analysis)
zone_dict = {
    'North': ['Delhi', 'Uttar Pradesh', 'Punjab', 'Haryana', 'Rajasthan', 'Himachal Pradesh', 'Jammu and Kashmir'],
    'South': ['Karnataka', 'Tamil Nadu', 'Kerala', 'Andhra Pradesh', 'Telangana'],
    'East': ['West Bengal', 'Odisha', 'Bihar', 'Jharkhand'],
    'West': ['Maharashtra', 'Gujarat', 'Goa', 'Madhya Pradesh']
}
```

```python
def get_zone(state):
    state = str(state).strip()
    for zone, states in zone_dict.items():
        if state in states:
            return zone
    return 'Other'

df['Zone'] = df['State'].apply(get_zone)
```
